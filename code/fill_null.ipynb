```python
### 지점소마다 인근 지점소에서 정보 가져올 수 있도록 최대 3개씩 연결 -> 근처에도 없을수도 잇으니까 일단 최대 3개씩하기 ###
### 근데 만약 너무 먼 곳이다 하면 그나마 가까운 1개만 해두기
aws_dict = {
    '강남' : ['서초', '송파', '광진'],
    '강동' : ['광진'], # 사실 강동은 구리 지점이나 하남 지점이 훨씬 가까운데 그건 못 쓰니까 일단 광진만
    '강북*' : ['성북', '도봉'],
    '강서' : ['양천'], # 마찬가지로 근처 서울 지점이 없어서 양천
    '관악' : ['관악(레)','남현', '금천'],
    '관악(레)' : ['관악', '남현', '금천'],
    '광진' : ['성동', '강남', '강동'],
    '구로' : ['양천', '영등포', '기상청'],
    '금천' : ['기상청', '관악', '영등포'],
    '기상청' : ['금천','현충원', '한강'],
    '남현' : ['관악', '관악(레)', '현충원'],
    '노원' : ['중랑', '도봉', '강북*'],
    '도봉' : ['강북*', '노원', '성북'],
    '동대문' : ['중랑', '노원', '성동'],
    '마포' : ['서대문', '한강', '영등포'],
    '북악산' : ['성북', '중구', '강북*'],
    '서대문' : ['마포', '중구', '한강'],
    '서초' : ['강남', '남현', '현충원'],
    '성동' : ['광진', '중구', '동대문'],
    '성북' : ['강북*', '동대문', '은평'],
    '송파' : ['강남', '서초', '광진'],
    '양천' : ['영등포', '구로', '강서'],
    '영등포' : ['기상청', '한강', '양천'],
    '용산' : ['중구', '현충원', '한강'],
    '은평' : ['성북', '서대문'],
    '중구' : ['서대문', '성동', '용산'],
    '중랑' : ['동대문', '노원'],
    '한강' : ['영등포', '용산', '기상청'],
    '현충원' : ['용산', '기상청', '한강'] 
}

### 채울 컬럼 미리 정의
columns_to_fill = [
    '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)',
    '평균풍속(m/s)', '평균습도(%)', '최저습도(%)', '최고습도(%)',
    '습도표준편차',
    '일_평균기온(°C)', '일_최저기온(°C)', '일_최고기온(°C)', '일_일강수량(mm)'
]

### 결측값을 인근 지점으로 보완
def fill_missing_by_neighbors(df, col, ref_dict):
    filled = []
    for i, row in df.iterrows():
        if pd.notnull(row[col]):
            filled.append(row[col])
        else:
            value = np.nan
            for neighbor in ref_dict.get(row['지점명'], []):
                # 동일한 날짜 & 인근 지점에서 대상 컬럼 값 찾음
                candidates = df.loc[
                    (df['날짜'] == row['날짜']) & (df['지점명'] == neighbor),
                    col
                ]
                # 해당 인근 지점에 값이 존재하고 NaN이 아니면, 그 값을 사용하고 반복 종료
                if not candidates.empty and pd.notnull(candidates.values[0]):
                    value = candidates.values[0]
                    break
            filled.append(value)
    df[col] = filled
    return df
```

```python
###### 결측치 보완을 위한 전처리 ######


### 20230101~ 데이터만 사용 (2022년 습도 누락 많음) 
df['날짜'] = pd.to_datetime(df['날짜'])
df = df[df['날짜'] >= pd.to_datetime("2023-01-01")]

### 벡터평균풍향(deg), 일_최대순간풍속(m/s), 일_최대순간풍속풍향(deg), 일_평균풍속(m/s) 컬럼 제거 ###
drop_col = ['벡터평균풍향(deg)', '일_최대순간풍속(m/s)','일_최대순간풍속풍향(deg)', '일_평균풍속(m/s)']
df = df.drop(drop_col, axis=1)

###  ['일_평균기온(°C)','일_최저기온(°C)', '일_최고기온(°C)', '일_일강수량(mm)']에 누락이 있거나 ['평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)'] 상호보완
# 서로 누락값 대체. 만약 '일_평균기온(°C)'이 누락이고 '평균기온(°C)'이 있으면 '일_평균기온(°C)'을 '평균기온(°C)'으로 대체하는 식으로 서로 채우기
# => 각 행(row) 단위로 결측값이 있는 경우, 그 행의 다른 열 값을 사용해서 채우는 방식
# (일별) → (시간별) 보완
df['일_평균기온(°C)'] = df['일_평균기온(°C)'].fillna(df['평균기온(°C)'])
df['일_최저기온(°C)'] = df['일_최저기온(°C)'].fillna(df['최저기온(°C)'])
df['일_최고기온(°C)'] = df['일_최고기온(°C)'].fillna(df['최고기온(°C)'])
df['일_일강수량(mm)'] = df['일_일강수량(mm)'].fillna(df['일강수량(mm)'])

# (시간별) → (일별) 보완 (선택적: 이후 제거할 예정이면 생략 가능)
df['평균기온(°C)'] = df['평균기온(°C)'].fillna(df['일_평균기온(°C)'])
df['최저기온(°C)'] = df['최저기온(°C)'].fillna(df['일_최저기온(°C)'])
df['최고기온(°C)'] = df['최고기온(°C)'].fillna(df['일_최고기온(°C)'])
df['일강수량(mm)'] = df['일강수량(mm)'].fillna(df['일_일강수량(mm)'])


### 관측 지점 대체해서 인근 지점 정보로 채우기 
for col in columns_to_fill:
    df = fill_missing_by_neighbors(df, col, aws_dict)

### 관악(레) 지점을 제외하는 것이 맞음 -> 애초에 과천시 소속이기도 하고 관악(레) 자체의 널값도 많음
df = df[df['지점명'] != '관악(레)'] # 기존 행에서 1277만 되어야함 왜냐면 len(df[df['지점명'] == '관악(레)'])이 1277이었음



df.info() # 20230101~20250630 : 912일
```

```python
### 지점별로 20230101~20250630 중에 아예 존재하지 않는 데이터 확인 

# 우리의 모든 날짜 정의
full_dates = pd.date_range(start="2023-01-01", end="2025-06-30", freq='D')

# 지점명 리스트
stations = df['지점명'].unique()

# 누락 정보 저장용
missing = []

for station in stations:
    # 해당 지점의 날짜 목록
    station_dates = df[df['지점명'] == station]['날짜'].dt.normalize().unique()
    station_dates = pd.to_datetime(station_dates)
    
    # 누락된 날짜 계산
    missing_dates = set(full_dates) - set(station_dates)
    
    for date in sorted(missing_dates):
        missing.append({'지점명': station, '누락날짜': date})

# 결과 DataFrame으로 만들기
missing_date = pd.DataFrame(missing)

# 지점별 누락일 수 확인
missing_summary = missing_date.groupby('지점명').size().reset_index(name='누락일수')

# 출력
print(missing_summary.sort_values(by='누락일수', ascending=False))
print(missing_date.head())
```

```python
### 아예 누락된 일자 데이터를 인근 지점정보로 가져오기 + 기존의 누락된 지점 정보(지점, 지점명, 주소, 위도, 경도, 행정구)가 제대로 채워져야함

def create_missing_rows(df, missing_df, aws_dict, meta):
    new_rows = []

    for _, row in missing_df.iterrows():
        station = row['지점명']
        date = row['누락날짜']

        # 인근 AWS 지점 목록 불러오기
        neighbors = aws_dict.get(station, [])
        
        for neighbor in neighbors:
            # 근처 지점에서 동일한 날짜 데이터 있는지 확인
            neighbor_row = df[(df['지점명'] == neighbor) & (df['날짜'] == date)]
            
            if not neighbor_row.empty:
                # 해당 행 복사
                copy_row = neighbor_row.iloc[0].copy()
                
                # 누락된 지점의 메타데이터 가져오기
                meta_row = meta[meta['지점명'] == station].iloc[0]

                # 누락된 지점 기준으로 메타 정보 덮어쓰기
                copy_row['지점명'] = meta_row['지점명']
                copy_row['지점'] = meta_row['AWS지점']
                copy_row['주소'] = meta_row['주소']
                copy_row['위도'] = meta_row['위도']
                copy_row['경도'] = meta_row['경도']
                copy_row['행정구'] = meta_row['행정구']

                # 날짜는 그대로 유지
                copy_row['날짜'] = date

                new_rows.append(copy_row)
                break  # 첫 번째로 채운 지점 기준으로만 사용

    return pd.DataFrame(new_rows)

# 누락된 행 생성
new_rows_df = create_missing_rows(df, missing_date, aws_dict, meta)

# 원본 데이터에 병합 후 정렬
df_filled = pd.concat([df, new_rows_df], ignore_index=True)\
              .sort_values(by=['지점명', '날짜'])\
              .reset_index(drop=True)

# 확인
print(f"추가된 행 수: {len(new_rows_df)}")
df_filled.info()
```
