```python
### 컬럼 제거 ###
# info 정보 기준 널값이 대다수인 경우 + 불필요한 컬럼('위생업태명', '소재지전화', ...)
# 우편번호는 상세주소 있으므로 제거
# 겹치는 정보 제거 : '영업상태구분코드', '상세영업상태코드', '영업상태명'
'''
'번호', '영업상태구분코드', '상세영업상태코드', '영업상태명', '인허가일자', '인허가취소일자', '휴업시작일자', '휴업종료일자', '재개업일자', '소재지전화', 
'소재지우편번호', '도로명우편번호', '건물지상층수', '건물지하층수', '사용시작지상층', '사용끝지상층', '사용시작지하층', 
'사용끝지하층','조건부허가신고사유', '조건부허가시작일자', '조건부허가종료일자', '건물소유구분명', 'Unnamed: 44', '위생업태명'
'''

data = data.drop(columns=['번호', '영업상태구분코드', '상세영업상태코드', '영업상태명', '인허가일자','인허가취소일자', '휴업시작일자', '휴업종료일자', '재개업일자', '소재지전화', 
'소재지우편번호', '도로명우편번호', '건물지상층수', '건물지하층수', '사용시작지상층', '사용끝지상층', '사용시작지하층', 
'사용끝지하층','조건부허가신고사유', '조건부허가시작일자', '조건부허가종료일자', '건물소유구분명', 'Unnamed: 44', '위생업태명'], axis=1)

data.columns
```

```python
data.info()
```

```python
data['폐업일자'] = pd.to_datetime(data['폐업일자'], errors='coerce')

# 폐업일자가 있고 & 2018-01-01 이전인 경우 제거
mask_drop = data['폐업일자'].notna() & (data['폐업일자'] < pd.Timestamp('2018-01-01'))

data = data[~mask_drop]
data.info()
```

```python
# 결측은 NaN 유지, 나머지만 int 변환
data['여성종사자수'] = data['여성종사자수'].astype('Int64')
data['남성종사자수'] = data['남성종사자수'].astype('Int64')
data['욕실수'] = data['욕실수'].astype('Int64')
```

```python
# 종사자수 계산 (결측은 0으로 처리)
data['종사자수'] = data[['여성종사자수', '남성종사자수']].fillna(0).sum(axis=1)
data.loc[(data['종사자수'].notna()) & (data['종사자수'] >= 1),['종사자수', '여성종사자수', '남성종사자수']]
```

```python
data['상세영업상태명'].value_counts() # 2018-01-01 이후 폐업한 곳이 469
```

```python
data.loc[data['폐업일자'].notna(), ['폐업일자', '사업장명', '상세영업상태명']] # 2018년 이후 폐업
```

```python
data['상세영업상태명'].value_counts()
```

```python
data['관리번호'].nunique()
```

```python
data.loc[
    data['소재지전체주소'].isna() | data['도로명전체주소'].isna(),
    ['사업장명','상세영업상태명', '소재지전체주소', '도로명전체주소']
]
```

```python
# 우성탕 폐업 -> 도로명 주소 찾아보면 '서울특별시 마포구 삼개로 38 (도화동, 도화3지구우성아파트)' 
# 도로명전체주소에서 널값이 소재지전체주소가 '서울특별시 마포구 도화동 82번지 우성상가 301호' 하나이니까 바꿔주기

mask = data['소재지전체주소'] == '서울특별시 마포구 도화동 82번지 우성상가 301호'
data.loc[mask, '도로명전체주소'] = '서울특별시 마포구 삼개로 38 (도화동, 도화3지구우성아파트)'

# 수정 결과 확인
data.loc[mask, ['소재지전체주소', '도로명전체주소']]
```

```python
data.loc[
    data['소재지전체주소'].isna() | data['도로명전체주소'].isna(),
    ['사업장명','상세영업상태명', '소재지전체주소', '도로명전체주소']
]
```

```python
data = data.drop(columns=['소재지전체주소'], axis=1)
data.info()
```
