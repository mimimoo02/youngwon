```python
### 행정동 추출 위해 ###

# 필수 라이브러리 import 및 기본 상수/엔드포인트 설정
# 주소 문자열 전처리 함수
# 외부 API(행안부/카카오) 호출 래퍼에 간단한 캐시 적용

import re, time, threading, requests, pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

# API 키
JUSO_KEY  = "devU01TX0FVVEgyMDI1MDgxMjE3MjIzMjExNjA2MjU=" # 도로표준화 때 사용 
KAKAO_KEY = "d86a315467c1c7369d876fd654b23cd0" # 행정동 추출할때 사용

# 엔드포인트 및 공통 HTTP 헤더
JUSO_URL   = "https://www.juso.go.kr/addrlink/addrLinkApi.do"  # 행안부 주소 표준화 API 
ADDR_URL   = "https://dapi.kakao.com/v2/local/search/address.json"  # 카카오 주소→좌표
REGION_URL = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json"  # 카카오 좌표→행정/법정동
HEADERS    = {"Authorization": f"KakaoAK {KAKAO_KEY}"}  # 카카오 인증 헤더

# 동시성 & 타임아웃
CONC_JUSO  = 12     # 행안부는 10~15 권장
CONC_KAKAO = 25     # 카카오는 20~30 권장
TIMEOUT = 7         # 각 요청의 타임아웃(초)
MAX_RETRY = 3       # 네트워크 오류 시 재시도 횟수(총 시도 = MAX_RETRY+1)
MAX_SHRINK = 12     # 뒤에서 1글자씩 줄이는 최대 시도 횟수 

## 문자열 가공
def norm_addr(s: str):
    """
    주소 문자열의 기본 정규화:
    - 입력이 문자열이 아니면 None 반환
    - 양 끝 공백 제거, 중복 공백을 단일 공백으로 축소
    - 결과가 빈 문자열이면 None
    """
    if not isinstance(s, str): return None
    s = re.sub(r"\s+", " ", s.strip())
    return s if s else None

def remove_trailing_bunji(s: str):
    """
    주소 문자열 끝이 정확히 '번지'로 끝나는 경우 그 부분만 제거
    - 카카오/행안부에서 '번지' 유무에 따라 매칭 실패하는 경우가 있어 후보 다양화 목적
    """
    if not isinstance(s, str): return None
    s = s.strip()
    return s[:-2].strip() if s.endswith("번지") else s

def cut_before_last_dash(s: str):
    """
    마지막 '-' 앞까지만 남김 (예: '67-138'→'67', '257-48번지'→'257번지')
    - 너무 세밀한 부번/호수 등은 제거하여 매칭률을 높이려는 전략
    """
    if not isinstance(s, str): return None
    m = re.search(r"^(.*?)(?:-\d+)(?:번지)?\s*$", s)
    return m.group(1).strip() if m else s

def progressive_shrink(s: str, max_try=MAX_SHRINK):
    """
    뒤에서 1글자씩 줄여가며 후보 생성 (너무 짧아지지 않게 컷오프)
    - 지나치게 짧아지는 것을 방지하기 위해 최소 길이(min_len) 설정
    - 경험적으로 '...구' 지점까지는 남기는 것이 지역 매칭에 유리해 컷오프에 반영
    """
    s0 = norm_addr(s)
    if not s0: return []
    # '구'의 위치를 찾고, 없으면 -1 → +1로 0이 됨; 최소 8자 보장
    min_len = max(8, s0.find("구")+1)  # '...구'까지는 남기자(대략적인 컷오프)
    out, cur = [], s0
    for _ in range(max_try):
        if len(cur) <= min_len: break
        cur = cur[:-1].rstrip()
        out.append(cur)
    return out

## 행안부 표준화 호출
std_cache, std_lock = {}, threading.Lock() # 같은 입력에 대한 중복 API 호출을 피하기 위해 스레드 세이프 캐시 사용
def juso_standardize(addr: str):
    """
    주소(지번/도로명 무관) -> 표준 도로명주소 or None 
    - 내부 캐시(std_cache)로 동일 주소 반복 호출 방지
    - 네트워크 오류 등 발생 시 재시도 (MAX_RETRY 기준)
    """
    
    if not addr: return addr, None
    a = norm_addr(addr)
    if not a: return addr, None
    with std_lock:
        if a in std_cache: return a, std_cache[a]
    params = {'confmKey': JUSO_KEY, 'currentPage':1, 'countPerPage':1,
              'keyword': a, 'resultType':'json'}
    out = None

    # 재시도 루프 (최대 MAX_RETRY+1회)
    for attempt in range(1, MAX_RETRY+2):
        try:
            r = requests.get(JUSO_URL, params=params, timeout=TIMEOUT); r.raise_for_status()
            j = r.json().get('results',{}).get('juso',[])
            out = j[0].get('roadAddr') if j else None # 최상위 후보 1건만 사용
            break
        except Exception:
            # 마지막 시도가 아니면 지수적 대기 후 재시도
            if attempt == MAX_RETRY+1: out = None
            else: time.sleep(0.5*attempt)
    
    # 캐시에 저장
    with std_lock:
        std_cache[a] = out
    return a, out
```

```python
### 표준도로명주소 생성
# 후보 주소 생성(정규화, ‘번지’ 제거, 하이픈 절단, 점진 축약) → 행안부 API 병렬 호출 → 첫 성공값으로 표준도로명주소 채움.

def candidate_sequence(row):
    # 원본 정규화 + '번지' 제거
    raw_road = remove_trailing_bunji(norm_addr(row.get('도로명전체주소')))
    raw_lot  = remove_trailing_bunji(norm_addr(row.get('소재지전체주소')))

    cand = []
    # 도로명 → 소재지 (둘 다 사용)
    for a in [raw_road, raw_lot]:
        if a and a not in cand: cand.append(a)

    # 소재지에서 마지막 '-' 앞까지만
    if raw_lot:
        cut = cut_before_last_dash(raw_lot)
        if cut and cut not in cand and cut != raw_lot:
            cand.append(cut)
        # 소재지 주소 뒤에서 1글자씩 줄이기
        for shr in progressive_shrink(raw_lot):
            if shr not in cand: cand.append(shr)

    return cand

# 표준도로명주소 컬럼 준비 (없으면 생성)
if '표준도로명주소' not in data.columns:
    data['표준도로명주소'] = None

# 표준주소가 비어 있는 행들만 대상 선정
need = data['표준도로명주소'].isna()

# 각 행별 후보열 시퀀스 생성
cseq = data.loc[need].apply(candidate_sequence, axis=1)

# 모든 후보 flatten → 유니크 : 후보 리스트를 평탄화(flatten) 후, 순서를 유지한 중복제거(dict.fromkeys 이용)
all_candidates = []
for lst in cseq: all_candidates.extend(lst)
uniq_candidates = list(dict.fromkeys([a for a in all_candidates if a]))  # 순서유지 dedupe
print("유니크 후보 수:", len(uniq_candidates))

# 병렬 표준화 호출 (후보별 1회)
t0 = time.time()
with ThreadPoolExecutor(max_workers=CONC_JUSO) as ex:
    futures = [ex.submit(juso_standardize, a) for a in uniq_candidates]
    for i, f in enumerate(as_completed(futures), 1):
        if i % 200 == 0 or i == len(futures):
            dt = time.time()-t0
            print(f"[JUSO 호출] {i}/{len(futures)} | {i/max(dt,1):.1f} req/s")

# 행별로 첫 성공값 채워넣기
filled_idx = []
for idx, lst in cseq.items():
    std = None
    for a in lst:
        key = norm_addr(a)
        std = std_cache.get(key) # 캐시에 저장된 표준주소 결과 조회
        if std: break
    if std:
        data.at[idx, '표준도로명주소'] = std
        filled_idx.append(idx)

print(f"이번 라운드 채워짐: {len(filled_idx)}")
print("표준도로명주소 결측:", int(data['표준도로명주소'].isna().sum()), "/", len(data))
```

```python
### 좌표(WGS84) 채우기 → 행정동
# 기존 좌표 (있으면) 5174→WGS84 변환, (없으면) 카카오 주소검색으로 좌표 보강 → 좌표로 행정동/법정동 역지오코딩(캐시+병렬).
# 좌표 있으면 변환 사용
X_COL, Y_COL = '좌표정보x(epsg5174)', '좌표정보y(epsg5174)'
if X_COL not in data.columns: data[X_COL] = None
if Y_COL not in data.columns: data[Y_COL] = None

def try_import_pyproj():
    """
    pyproj(좌표계 변환 라이브러리) 사용 가능 여부 확인.
    - 사용할 수 있으면 5174 → 4326(WGS84) 변환 Transformer 반환
    - 설치/환경 문제 시 None 반환하여 후속 로직이 안전하게 건너뛰도록 함
    """
    try:
        from pyproj import Transformer
        return Transformer.from_crs(5174, 4326, always_xy=True)
    except Exception:
        return None

_transformer = try_import_pyproj()

def tm5174_to_wgs84(x, y):
    """
    EPSG:5174 (TM중부원점) 좌표를 WGS84 경위도(lon, lat)로 변환.
    - 변환기(_transformer)가 없거나 입력이 결측이면 (None, None) 반환
    - 예외 발생 시에도 (None, None)으로 안전하게 처리
    """
    if pd.isna(x) or pd.isna(y) or _transformer is None: return (None, None)
    try:
        lon, lat = _transformer.transform(float(x), float(y))
        return (lon, lat)
    except Exception:
        return (None, None)

# 기존 좌표 → 경위도 변환 적용
xy0 = data[[X_COL, Y_COL]].apply(lambda r: tm5174_to_wgs84(r[X_COL], r[Y_COL]), axis=1)
data['lon'] = [p[0] for p in xy0]
data['lat'] = [p[1] for p in xy0]
print("기존좌표 변환 후 결측(lon/lat):", int((data['lon'].isna()|data['lat'].isna()).sum()))

# 좌표 없는 행만 카카오 주소검색(표준도로명주소로)
addr_cache, addr_lock = {}, threading.Lock()
def kakao_geocode(addr: str):
    """
    카카오 주소검색 API로 주소 → (x=lon, y=lat) 좌표를 획득
    - 내부 캐시(addr_cache)와 락으로 중복호출/경쟁상황 방지
    - 재시도 및 backoff 적용
    """
    if not isinstance(addr, str) or not addr.strip(): return (None, None)
    a = norm_addr(addr)
    with addr_lock:
        if a in addr_cache: return addr_cache[a]
    out = (None, None)
    for attempt in range(1, MAX_RETRY+2):
        try:
            r = requests.get(ADDR_URL, headers=HEADERS, params={"query": a}, timeout=TIMEOUT)
            r.raise_for_status()
            docs = r.json().get("documents", [])
            out = (float(docs[0]['x']), float(docs[0]['y'])) if docs else (None, None)
            break
        except Exception:
            if attempt == MAX_RETRY+1: out = (None, None)
            else: time.sleep(0.4*attempt)
    with addr_lock:
        addr_cache[a] = out
    return out

# lon/lat 결측인 행만 대상으로 고유 표준주소 목록을 생성하여 병렬 지오코딩
need_geo = data['lon'].isna() | data['lat'].isna()
uniq_std = data.loc[need_geo, '표준도로명주소'].dropna().unique().tolist()

t2 = time.time()
with ThreadPoolExecutor(max_workers=CONC_KAKAO) as ex:
    futures = [ex.submit(kakao_geocode, a) for a in uniq_std]
    for i, f in enumerate(as_completed(futures), 1):
        if i % 200 == 0 or i == len(futures):
            dt = time.time()-t2
            print(f"[카카오 지오코딩] {i}/{len(uniq_std)} | {i/max(dt,1):.1f} req/s")

# 지오코딩 결과를 행 단위로 매핑하여 lon/lat 채우기
data.loc[need_geo, 'lon'] = data.loc[need_geo, '표준도로명주소'].map(
    lambda a: addr_cache.get(norm_addr(a),(None,None))[0]
).astype('float64').values
data.loc[need_geo, 'lat'] = data.loc[need_geo, '표준도로명주소'].map(
    lambda a: addr_cache.get(norm_addr(a),(None,None))[1]
).astype('float64').values
print("최종 좌표 결측:", int((data['lon'].isna() | data['lat'].isna()).sum()))

# 좌표 -> 행정동/법정동
rc_cache, rc_lock = {}, threading.Lock()
def coord2region(lon, lat):
    """
    카카오 coord2regioncode API를 사용해 좌표 → 행정동/법정동 메타데이터를 조회
    - region_type 'H' (행정동), 'B' (법정동) 각각의 3단계 이름 및 코드 반환
    - 네트워크 오류 시 재시도, 그래도 실패하면 None 필드 유지
    """
    if pd.isna(lon) or pd.isna(lat):
        return {"행정동":None,"행정동코드":None,"법정동명":None,"법정동코드":None}
    key = (float(lon), float(lat))
    with rc_lock:
        if key in rc_cache: return rc_cache[key]
    out = {"행정동":None,"행정동코드":None,"법정동명":None,"법정동코드":None}
    for attempt in range(1, MAX_RETRY+2):
        try:
            r = requests.get(REGION_URL, headers=HEADERS, params={"x": key[0], "y": key[1]}, timeout=TIMEOUT)
            r.raise_for_status()
            docs = r.json().get("documents", [])
            h = next((d for d in docs if d.get('region_type')=='H'), None)
            b = next((d for d in docs if d.get('region_type')=='B'), None)
            out = {
                "행정동": (h or {}).get("region_3depth_name"),
                "행정동코드": (h or {}).get("code"),
                "법정동명": (b or {}).get("region_3depth_name"),
                "법정동코드": (b or {}).get("code"),
            }
            break
        except Exception:
            if attempt == MAX_RETRY+1: pass
            else: time.sleep(0.4*attempt)
    with rc_lock:
        rc_cache[key] = out
    return out

# 중복 좌표에 대한 API 호출을 줄이기 위해 고유 (lon, lat) 페어에 대해서만 병렬 조회
pairs = pd.Series(list(zip(data['lon'], data['lat']))).dropna().unique().tolist()
t3 = time.time()
with ThreadPoolExecutor(max_workers=CONC_KAKAO) as ex:
    futures = [ex.submit(coord2region, *p) for p in pairs]
    for i, f in enumerate(as_completed(futures), 1):
        if i % 300 == 0 or i == len(futures):
            dt = time.time()-t3
            print(f"[행정동 조회] {i}/{len(pairs)} | {i/max(dt,1):.1f} req/s")

# 조회 결과를 기존 데이터프레임 순서에 맞춰 병합
tmp = data[['lon','lat']].apply(lambda r: rc_cache.get((r['lon'], r['lat']),
                    {"행정동":None,"행정동코드":None,"법정동명":None,"법정동코드":None}), axis=1)
tmp = pd.json_normalize(tmp)
data[['행정동','행정동코드','법정동명','법정동코드']] = tmp[['행정동','행정동코드','법정동명','법정동코드']].values

print("총 건수:", len(data), " | 행정동 결측:", int(data['행정동'].isna().sum()))
data[['소재지전체주소','도로명전체주소','표준도로명주소','lon','lat','행정동']].head(10)
```

```python
### 행정동 기반 목용장업 데이터에 물재생센터 정보 병합 ###
# 먼저 '행정동' 이름 기준으로 병합 시도 (validate="m:1" 안전장치)
# 만약 매핑에서 H_DNG_NM 중복 때문에 실패하거나 안전하지 않으면,
#    - 중복이지만 처리센터가 모두 동일하면 안전 축약 후 이름 기준 병합
#    - 그게 아니면 "코드 기준 폴백":
#       df: 행정동명 정정(신설/용두→용신, 면목제3.8→면목3.8)
#       df: 행정동코드_trim = 행정동코드 // 100 (뒤 '00' 제거)
#       mapping: 행정동코드는 이미 8자리(trim) → 숫자형 변환만
#       df에서 '용신동'인 행의 행정동코드_trim을 매핑의 용신동 코드로 덮어쓰기
#       df['행정동코드_trim'] ↔ mapping['행정동코드'] 로 left join


df = pd.read_csv('../data/processed/인허가_목욕장업_행정동추출.csv', encoding='utf-8-sig')
mapping_df = pd.read_excel('../data/raw/행정동16년 기준 매핑.xlsx', engine='openpyxl')


# 행정동명 정정
# 신설동, 용두동 -> 용신동
# 면목제3.8동 -> 면목3.8동
name_fix = {'신설동': '용신동', 
            '용두동': '용신동', 
            '면목제3.8동': '면목3.8동'}
df['행정동'] = df['행정동'].replace(name_fix)

# '행정동' 컬럼 기반 병합 우선 시도 
def try_merge_by_name(df_in, map_in):
    """
    H_DNG_NM이 유일하면 이름 기준 병합을 시도하고 결과를 반환.
    유일하지 않다면 None을 반환.
    """
    if map_in['H_DNG_NM'].duplicated().any():
        return None  # 유일하지 않음 → 폴백 필요

    # 병합 (행정동 이름 기준)
    merged_name = df_in.merge(
        map_in[['H_DNG_NM', '처리센터']],
        how='left',
        left_on='행정동',
        right_on='H_DNG_NM',
        validate='m:1'  # 왼쪽 many, 오른쪽 one
    ).rename(columns={'처리센터':'물재생센터'}).drop(columns=['H_DNG_NM'])
    return merged_name

# 일단 시도
merged = try_merge_by_name(df, mapping_df)

# 이름 중복이 있다면: 중복 안전 축약 or 코드 폴백
if merged is None:
    # 중복 진단: 같은 동 이름에 서로 다른 처리센터가 있는지 확인
    g = mapping_df.groupby('H_DNG_NM')['처리센터'].nunique().reset_index(name='센터_고유값_수')
    conflict_keys = g.loc[g['센터_고유값_수'] > 1, 'H_DNG_NM'].tolist()

    if len(conflict_keys) == 0:
        # 처리센터가 실제로는 동일 → 완전 중복(동일 동+센터)이라면 안전 축약 가능
        map_dedup_safe = mapping_df.drop_duplicates(subset=['H_DNG_NM', '처리센터'], keep='first')
        # 여전히 H_DNG_NM 중복이 남는지 확인
        if map_dedup_safe['H_DNG_NM'].duplicated().any():
            # 이론상 거의 없지만, 혹시 남아있으면 코드 폴백으로 전환
            need_fallback = True
        else:
            need_fallback = False
            merged = try_merge_by_name(df, map_dedup_safe)
    else:
        # 같은 동 이름에 처리센터가 서로 다름 → 이름만으론 모호 → 코드 폴백
        need_fallback = True

    # 코드 폴백 (필요한 경우만)
    if merged is None and need_fallback:
        # df 쪽 코드 정규화: 뒤 2자리 제거한 8자리 trim 생성 <- 매핑 데이터의 행정동코드가 이미 8자리로 되어 있기때문에 필요
        df['행정동코드'] = pd.to_numeric(df['행정동코드'], errors='coerce').astype('Int64')
        df['행정동코드_trim'] = (df['행정동코드'] // 100).astype('Int64')

        # mapping 쪽은 이미 8자리 → 숫자형만 보정
        mapping_df['행정동코드'] = pd.to_numeric(mapping_df['행정동코드'], errors='coerce').astype('Int64')

        # 용신동 코드(8자리) 추출
        ys_code = mapping_df.loc[mapping_df['H_DNG_NM'] == '용신동', '행정동코드']
        if ys_code.empty:
            raise ValueError("매핑 파일에서 '용신동'의 행정동코드를 찾지 못했습니다.")
        ys_code = ys_code.iloc[0]

        # df에서 '용신동'으로 정정된 모든 행에 용신동 코드 적용
        df.loc[df['행정동'].eq('용신동'), '행정동코드_trim'] = ys_code

        # 코드 기준 병합
        merged = df.merge(
            mapping_df[['행정동코드', '처리센터']].drop_duplicates(subset=['행정동코드']),
            how='left',
            left_on='행정동코드_trim',
            right_on='행정동코드'
        ).rename(columns={'처리센터':'물재생센터'})

print(f"[요약] 전체 행: {len(merged):,d}")
print(f"[요약] 매핑 성공: {(~merged['물재생센터'].isna()).sum():,d}")
print(f"[요약] 매핑 실패: {merged['물재생센터'].isna().sum():,d}")

merged.to_csv('../data/processed/인허가_목욕장업_물재생센터매핑_결과.csv',
              index=False, encoding='utf-8-sig')
print("[저장 완료] ../data/processed/인허가_목욕장업_물재생센터매핑_결과.csv")
```

```python
# 2017-01-01 ~ 2025-06-30 기간에 대해 물재생센터 4개 × 매일 1행으로 구성된 일자별 데이터
"""
2017-01-01 ~ 2025-06-30 기간에 대해
물재생센터 4개 x 매일 1행이 존재하는 일자별 데이터를 생성

- 개업일자 정보 없음 → 모든 시설은 2017-01-01부터 존재한다고 가정
- 폐업일자 존재 시 → 폐업일 '전날'까지 영업(폐업일 당일은 영업 X)
- 날짜×센터의 모든 조합을 '풀그리드'로 만든 뒤, 누락은 0/기본값으로 채움

입출력
- 입력:  ../data/processed/인허가_목욕장업_일자별정리_준비.csv
- 출력:  ../data/processed/인허가_목욕장업_일자별_물재생센터_영업현황_20170101_20250630.csv
"""

import pandas as pd
import numpy as np
from itertools import product

# 기간 설정
START = pd.Timestamp("2017-01-01")
END = pd.Timestamp("2025-06-30")

# 날짜×센터 풀그리드 생성 시 사용할 '일 단위' 날짜 인덱스
date_index = pd.date_range(START, END, freq="D")

src_path = "../data/processed/인허가_목욕장업_일자별정리_준비.csv"
df = pd.read_csv(src_path, encoding="utf-8-sig")

# 수치형 변환: 면적은 숫자형으로 통일 (문자 → NaN 허용)
df["소재지면적"] = pd.to_numeric(df["소재지면적"], errors="coerce")
# 날짜형 변환: 문자열 → datetime (변환 불가/빈 값은 NaT)
df["폐업일자"] = pd.to_datetime(df["폐업일자"], errors="coerce")

# 물재생센터 목록 추출 (요구사항: 최종적으로 날짜마다 센터 4개 행이 나와야 함)
centers = df["물재생센터"].unique().tolist()

# 영업 기간 계산 : 개업일자 정보가 없으므로 모든 시설이 START부터 존재한다고 가정
df["영업시작일"] = START

# 폐업일자가 있으면 폐업일 전날까지 영업으로 간주
#  - 예: 폐업일=2020-01-10 → 영업종료일_포함 = 2020-01-09
df["영업종료일_포함"] = df["폐업일자"] - pd.Timedelta(days=1)
# 폐업일자 NaT(=미폐업)인 시설은 END까지 영업한다고 간주
df["영업종료일_포함"] = df["영업종료일_포함"].fillna(END)
# 영업 종료일이 분석 종료일을 넘어가면 END로 클리핑
df["영업종료일_포함"] = df["영업종료일_포함"].where(df["영업종료일_포함"] <= END, END)
# 종료일 < 시작일인 레코드는 분석 기간과 겹치지 않으므로 제거
df = df[df["영업종료일_포함"] >= df["영업시작일"]].copy()

# 시설 단위 → 일자 단위 확장 
# 각 시설 레코드마다 [영업시작일, 영업종료일_포함] 구간의 모든 날짜를 생성하여
# 시설 속성(면적, 발한실여부, 다중이용업소여부 등)을 일자별로 복제
def make_daily_rows(row):
    # 시설의 실제 운영 기간과 분석 기간이 겹치는 부분만 날짜 생성
    dts = pd.date_range(max(row["영업시작일"], START),
                        min(row["영업종료일_포함"], END),
                        freq="D")
    if len(dts) == 0:
        # 겹치는 날짜가 전혀 없을 경우 빈 프레임 반환
        return pd.DataFrame(columns=[
            "date", "물재생센터", "개방서비스명",
            "소재지면적", "발한실여부", "다중이용업소여부", "시설_운영"
        ])
    # 운영 중인 기간의 각 날짜별 행 생성
    return pd.DataFrame({
        "date": dts,
        "물재생센터": row["물재생센터"],
        "개방서비스명": row["개방서비스명"],
        "소재지면적": row["소재지면적"],
        "발한실여부": row["발한실여부"],
        "다중이용업소여부": row["다중이용업소여부"],
        # 이후 집계에서 '운영 시설 수'를 계산하기 위해
        "시설_운영": 1
    })

# 모든 시설 레코드에 대해 일자 단위로 확장
daily_list = [make_daily_rows(r) for _, r in df.iterrows()]

# 빈 리스트 방지: concat 전에 조건 처리
fac_daily = (
    pd.concat(daily_list, ignore_index=True)
    if len(daily_list) > 0 else
    pd.DataFrame(columns=[
        "date", "물재생센터", "개방서비스명",
        "소재지면적", "발한실여부", "다중이용업소여부", "시설_운영"
    ])
)

# 날짜×센터 집계
# 동일한 날짜/센터 내에서 여러 시설(행)이 존재할 수 있으므로
#  - 시설수_운영: 해당 일자에 운영 중인 시설 수 (시설_운영 합)
#  - 총소재지면적_운영: 운영 시설들의 면적 합
#  - 발한실_운영: 운영 시설 중 발한실여부 == "Y" 개수
#  - 다중이용_운영: 운영 시설 중 다중이용업소여부 == "Y" 개수
# 이후 시설수_운영 > 0이면 영업여부_센터 = 1, 아니면 0
if not fac_daily.empty:
    grp = fac_daily.groupby(["date", "물재생센터"], as_index=False).agg(
        시설수_운영=("시설_운영", "sum"),
        총소재지면적_운영=("소재지면적", "sum"),
        발한실_운영=("발한실여부", lambda s: (s == "Y").sum()),
        다중이용_운영=("다중이용업소여부", lambda s: (s == "Y").sum()),
    )
    grp["영업여부_센터"] = (grp["시설수_운영"] > 0).astype(int)

    # 개방서비스명
    mode_by_center = (
        fac_daily.groupby("물재생센터")["개방서비스명"]
        .agg(lambda s: s.mode().iat[0] if not s.mode().empty else s.iloc[0])
    )
    grp = grp.merge(mode_by_center.rename("개방서비스명"), on="물재생센터", how="left")
else:
    # 예외 처리 : 확장 결과가 비었을 때의 스키마 유지
    grp = pd.DataFrame(columns=[
        "date", "물재생센터", "시설수_운영", "총소재지면적_운영",
        "발한실_운영", "다중이용_운영", "영업여부_센터", "개방서비스명"
    ])

# 날짜×센터 풀그리드 결합
# 모든 날짜(date_index) × 모든 센터(centers) 조합을 만든 뒤,
# 위 집계(grp)를 LEFT JOIN → 누락된 조합은 0/기본값으로 채움
full_grid = pd.DataFrame(list(product(date_index, centers)), columns=["date", "물재생센터"])
daily_center = full_grid.merge(grp, on=["date", "물재생센터"], how="left")

# 결측치 처리 규칙:
#  - 운영 관련 수치(시설수_운영, 총소재지면적_운영, 발한실_운영, 다중이용_운영, 영업여부_센터) → 0
#  - 개방서비스명 → 센터별/전체 최빈값(없으면 '목욕장업')
default_service = (
    fac_daily["개방서비스명"].mode().iat[0]
    if not fac_daily.empty and not fac_daily["개방서비스명"].mode().empty
    else "목욕장업"
)
for col in ["시설수_운영", "총소재지면적_운영", "발한실_운영", "다중이용_운영", "영업여부_센터"]:
    daily_center[col] = daily_center[col].fillna(0)
daily_center["개방서비스명"] = daily_center["개방서비스명"].fillna(default_service)

# 최종 컬럼 순서로 정리하고, 날짜/센터 기준 오름차순 정렬
daily_center = daily_center[[
    "date", "물재생센터", "개방서비스명",
    "영업여부_센터", "시설수_운영", "총소재지면적_운영",
    "발한실_운영", "다중이용_운영"
]].sort_values(["date", "물재생센터"], ignore_index=True)

out_path = "../data/processed/인허가_목욕장업_일자별_물재생센터_영업현황_20170101_20250630.csv"
daily_center.to_csv(out_path, index=False, encoding="utf-8-sig")

print(f"[저장 완료] {out_path}")
daily_center.head(10)
```

```python
# =========================================================
# ✅ 공통 컬럼 삭제 + 폐업일자(< 2017-01-01) 필터 일괄 적용 (주석 풍부 버전)
# 대상 변수: data_pool, data_laundry, data_sports, data_gym, data_beauty, data_hair
# - 각 DF에 대해:
#   (1) 불필요 공통 컬럼 삭제 (존재하는 것만 안전하게 drop)
#   (2) '폐업일자'가 있고 & 2017-01-01 이전인 행 제거
# 결과 변수: final_pool, final_laundry, final_sports, final_gym, final_beauty, final_hair
# ---------------------------------------------------------
# ⚠️ '인허가일자'는 아래 DROP 리스트에 포함 → 정말 필요하면 PROTECT에 넣어서 보호하세요.
# =========================================================

import pandas as pd

# (A) 제거 대상 공통 컬럼 목록 — 네가 지정한 컬럼들 반영
DROP = [
    '번호','개방서비스아이디','개방자치단체코드','관리번호',
    '인허가일자','인허가취소일자',
    '영업상태구분코드','영업상태명','상세영업상태코드',
    '휴업시작일자','휴업종료일자','재개업일자',
    '소재지전화','소재지우편번호','도로명우편번호',
    '최종수정시점','데이터갱신구분','데이터갱신일자',
    '업태구분명','법인명'
    # 주의: 파일마다 'Unnamed: 28/37/40/44' 등 임시 컬럼이 달라서
    #       아래에서 'Unnamed:'로 시작하는 컬럼은 자동으로 함께 제거합니다.
]

# (B) 보호 컬럼 — 여기 넣으면 DROP에 있더라도 제거하지 않음
#     (예: 일자 집계가 필요해서 '인허가일자'를 살리고 싶다면 아래처럼 설정)
# PROTECT = ['인허가일자']
PROTECT = []

def clean_and_filter(df: pd.DataFrame, cutoff: str = '2017-01-01') -> pd.DataFrame:
    """
    하나의 DataFrame에 대해 '공통 컬럼 삭제' + '폐업일자 기준 필터'를 적용하여 사본을 반환.
    - df: 원본 DataFrame (원본은 보존)
    - cutoff: 'YYYY-MM-DD' 문자열. 이 날짜 이전의 폐업건은 제거.
    """
    out = df.copy()

    # 1) 삭제할 컬럼 목록 구성 ----------------------------------------
    #    - 사용자가 지정한 DROP 중 실제로 존재하는 컬럼만 선택
    #    - PROTECT에 있는 컬럼은 제거 대상에서 제외
    #    - 'Unnamed:'로 시작하는 임시 컬럼은 모두 자동 포함
    auto_unnamed = [c for c in out.columns if str(c).strip().startswith('Unnamed:')]
    drop_cols = [c for c in DROP if (c in out.columns) and (c not in PROTECT)] + auto_unnamed

    # 중복 제거(순서 유지) 후 안전 drop (없는 컬럼은 무시)
    out = out.drop(columns=list(dict.fromkeys(drop_cols)), errors='ignore')

    # 2) 폐업일자 필터 -----------------------------------------------
    #    - '폐업일자'가 존재하는 경우에만 적용
    #    - 문자열/숫자/빈값 등 다양한 형식이 들어올 수 있으므로 to_datetime(errors='coerce')로 NaT 처리
    #    - "폐업일자 값이 있고" AND "cutoff 이전" → 제거
    if '폐업일자' in out.columns:
        close_series = pd.to_datetime(out['폐업일자'], errors='coerce')   # 잘못된 값은 NaT
        cutoff_ts = pd.Timestamp(cutoff)

        # 제거할 행 마스크: (값이 존재) & (cutoff 이전)
        mask_drop = close_series.notna() & (close_series < cutoff_ts)

        # 실제 제거
        out = out.loc[~mask_drop].reset_index(drop=True)

    return out

# (C) 현재 커널에 올라와 있는 대상들만 자동 처리 --------------------
targets = ['data_pool','data_laundry','data_sports','data_gym','data_beauty','data_hair']

for name in targets:
    if name in globals():
        # 결과 변수명 만들기: data_pool → final_pool
        suffix = name.split('_', 1)[-1]
        out_name = f'final_{suffix}'

        # 처리 + 결과 바인딩
        before_shape = globals()[name].shape
        globals()[out_name] = clean_and_filter(globals()[name], cutoff='2017-01-01')
        after_shape = globals()[out_name].shape

        # 진행 로그: 몇 행 줄었는지 빠르게 확인
        print(f"[{name}] {before_shape} → {after_shape} | 결과: {out_name}")

# (D) (선택) 결과 크기만 한 번 더 요약 출력 -------------------------
for var in ['final_pool','final_laundry','final_sports','final_gym','final_beauty','final_hair']:
    if var in globals():
        print(f"{var}: {globals()[var].shape}")
```

```python
import re
import time
import threading
import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

# --- 1. API 설정 및 전역 변수 ---
try:
    from pyproj import Transformer
    _transformer_5174_to_wgs84 = Transformer.from_crs(5174, 4326, always_xy=True)
except ImportError:
    _transformer_5174_to_wgs84 = None
    print("Warning: 'pyproj' library not found. EPSG:5174 coordinate conversion will be skipped.")

JUSO_KEY = "devU01TX0FVVEgyMDI1MDgxMjE3MjIzMjExNjA2MjU="
KAKAO_KEY = "d86a315467c1c7369d876fd654b23cd0"
JUSO_URL = "https://www.juso.go.kr/addrlink/addrLinkApi.do"
ADDR_URL = "https://dapi.kakao.com/v2/local/search/address.json"
REGION_URL = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json"
HEADERS = {"Authorization": f"KakaoAK {KAKAO_KEY}"}
CONC_JUSO, CONC_KAKAO, TIMEOUT, MAX_RETRY = 12, 25, 10, 5
MAX_SHRINK = 12
_std_cache, _std_lock = {}, threading.Lock()
_addr_cache, _addr_lock = {}, threading.Lock()
_rc_cache, _rc_lock = {}, threading.Lock()


# --- 2. 메인 처리 함수 ---
def process_addresses(df: pd.DataFrame) -> pd.DataFrame:
    data = df.copy()

    # (내부 헬퍼 및 API 호출 함수들은 이전과 동일하게 유지)
    def norm_addr(s: str):
        if not isinstance(s, str): return None
        s = re.sub(r"\s+", " ", s.strip())
        return s if s else None
    def remove_trailing_bunji(s: str):
        s = norm_addr(s)
        return s[:-2].strip() if s and s.endswith("번지") else s
    def cut_before_last_dash(s: str):
        s = norm_addr(s)
        if not s: return None
        m = re.search(r"^(.*?)(?:-\d+)(?:번지)?\s*$", s)
        return m.group(1).strip() if m else s
    def progressive_shrink(s: str, max_try=MAX_SHRINK):
        s0 = norm_addr(s)
        if not s0: return []
        min_len = max(8, s0.find("구") + 1)
        out, cur = [], s0
        for _ in range(max_try):
            if len(cur) <= min_len: break
            cur = cur[:-1].rstrip()
            if cur: out.append(cur)
        return out
    def candidate_sequence(row):
        raw_road = remove_trailing_bunji(row.get('도로명전체주소'))
        raw_lot = remove_trailing_bunji(row.get('소재지전체주소'))
        cand = []
        for a in [raw_road, raw_lot]:
            if a and a not in cand: cand.append(a)
        if raw_lot:
            cut = cut_before_last_dash(raw_lot)
            if cut and cut != raw_lot and cut not in cand: cand.append(cut)
            for shr in progressive_shrink(raw_lot):
                if shr not in cand: cand.append(shr)
        return cand
    def juso_standardize(addr: str):
        a = norm_addr(addr)
        if not a: return
        with _std_lock:
            if a in _std_cache: return
        params = {'confmKey': JUSO_KEY, 'currentPage': 1, 'countPerPage': 1, 'keyword': a, 'resultType': 'json'}
        out = None
        for attempt in range(1, MAX_RETRY + 2):
            try:
                r = requests.get(JUSO_URL, params=params, timeout=TIMEOUT); r.raise_for_status()
                j = r.json().get('results', {}).get('juso', [])
                out = j[0].get('roadAddr') if j else None
                break
            except Exception:
                if attempt > MAX_RETRY: out = None
                else: time.sleep(0.5 * attempt)
        with _std_lock: _std_cache[a] = out
    def kakao_geocode(addr: str):
        a = norm_addr(addr)
        if not a: return
        with _addr_lock:
            if a in _addr_cache: return
        out = (None, None)
        for attempt in range(1, MAX_RETRY + 2):
            try:
                r = requests.get(ADDR_URL, headers=HEADERS, params={"query": a}, timeout=TIMEOUT); r.raise_for_status()
                docs = r.json().get("documents", [])
                out = (float(docs[0]['x']), float(docs[0]['y'])) if docs else (None, None)
                break
            except Exception:
                if attempt > MAX_RETRY: out = (None, None)
                else: time.sleep(0.4 * attempt)
        with _addr_lock: _addr_cache[a] = out
    def coord2region(lon, lat):
        default_out = {"행정동": None, "행정동코드": None, "법정동명": None, "법정동코드": None}
        if pd.isna(lon) or pd.isna(lat): return
        key = (float(lon), float(lat))
        with _rc_lock:
            if key in _rc_cache: return
        out = default_out.copy()
        for attempt in range(1, MAX_RETRY + 2):
            try:
                r = requests.get(REGION_URL, headers=HEADERS, params={"x": key[0], "y": key[1]}, timeout=TIMEOUT); r.raise_for_status()
                docs = r.json().get("documents", [])
                h = next((d for d in docs if d.get('region_type') == 'H'), None)
                b = next((d for d in docs if d.get('region_type') == 'B'), None)
                out = {
                    "행정동": (h or {}).get("region_3depth_name"), "행정동코드": (h or {}).get("code"),
                    "법정동명": (b or {}).get("region_3depth_name"), "법정동코드": (b or {}).get("code"),
                }
                break
            except Exception:
                if attempt > MAX_RETRY: pass
                else: time.sleep(0.4 * attempt)
        with _rc_lock: _rc_cache[key] = out

    # --- 단계 1: 주소 정제 및 상호 보완 ---
    print("--- 1. 주소 정제 시작 ---")
    data['소재지전체주소'] = data['소재지전체주소'].fillna(data['도로명전체주소'])
    data['도로명전체주소'] = data['도로명전체주소'].fillna(data['소재지전체주소'])
    for col in ['소재지전체주소', '도로명전체주소']:
        if col in data and data[col].dtype == 'object':
            data[col] = data[col].str.split().str[:4].str.join(' ').str.split(',').str.get(0).str.strip()

    # --- 단계 2: 표준도로명주소 생성 ---
    print("\n--- 2. 표준도로명주소 생성 시작 ---")
    if '표준도로명주소' not in data.columns: data['표준도로명주소'] = None
    need_std = data[data['표준도로명주소'].isna()]
    if not need_std.empty:
        cseq = need_std.apply(candidate_sequence, axis=1)
        uniq_candidates = list(dict.fromkeys(item for sublist in cseq for item in sublist if item))
        print(f"총 {len(uniq_candidates)}개의 고유 주소 후보로 표준화 API 호출...")
        with ThreadPoolExecutor(max_workers=CONC_JUSO) as ex:
            list(ex.map(juso_standardize, uniq_candidates))
        for idx, lst in cseq.items():
            for a in lst:
                std = _std_cache.get(norm_addr(a))
                if std:
                    data.at[idx, '표준도로명주소'] = std
                    break
    print(f"표준화 완료. 결측 수: {data['표준도로명주소'].isna().sum()}/{len(data)}")

    # --- 단계 3: 좌표 변환 및 보강 ---
    print("\n--- 3. 좌표 변환 및 보강 시작 ---")
    X_COL, Y_COL = '좌표정보x(epsg5174)', '좌표정보y(epsg5174)'
    if X_COL not in data.columns: data[X_COL] = None
    if Y_COL not in data.columns: data[Y_COL] = None
    if 'lon' not in data.columns: data['lon'] = None
    if 'lat' not in data.columns: data['lat'] = None
    if _transformer_5174_to_wgs84:
        valid_tm_coords_idx = data[pd.notna(data[X_COL]) & pd.notna(data[Y_COL])].index
        coords = data.loc[valid_tm_coords_idx].apply(
            lambda r: _transformer_5174_to_wgs84.transform(float(r[X_COL]), float(r[Y_COL])), axis=1)
        if not coords.empty:
            data.loc[coords.index, 'lon'] = [c[0] for c in coords]
            data.loc[coords.index, 'lat'] = [c[1] for c in coords]
    need_geo = data['lon'].isna()
    if need_geo.any():
        uniq_std_to_geo = data.loc[need_geo, '표준도로명주소'].dropna().unique()
        print(f"총 {len(uniq_std_to_geo)}개의 고유 표준주소로 지오코딩 API 호출...")
        with ThreadPoolExecutor(max_workers=CONC_KAKAO) as ex:
            list(ex.map(kakao_geocode, uniq_std_to_geo))
        geo_results = data.loc[need_geo, '표준도로명주소'].map(lambda a: _addr_cache.get(norm_addr(a), (None, None)))
        if not geo_results.empty:
            data.loc[need_geo, 'lon'] = [r[0] for r in geo_results]
            data.loc[need_geo, 'lat'] = [r[1] for r in geo_results]
    print(f"좌표 보강 완료. 최종 좌표 결측 수: {(data['lon'].isna() | data['lat'].isna()).sum()}/{len(data)}")

    # --- 단계 4: 좌표로 행정동 추출 ---
    print("\n--- 4. 행정동 추출 시작 ---")
    valid_coords = data[['lon', 'lat']].dropna().drop_duplicates()
    pairs = list(valid_coords.itertuples(index=False, name=None))
    if pairs:
        print(f"총 {len(pairs)}개의 고유 좌표로 역지오코딩 API 호출...")
        with ThreadPoolExecutor(max_workers=CONC_KAKAO) as ex:
            list(ex.map(lambda p: coord2region(*p), pairs))

    # ★★★ 수정: 최종 병합 방식을 원본과 동일한 .values 방식으로 변경 ★★★
    default_region = {"행정동": None, "행정동코드": None, "법정동명": None, "법정동코드": None}
    tmp = data.apply(lambda r: _rc_cache.get((r['lon'], r['lat']), default_region) if pd.notna(r['lon']) else default_region, axis=1)
    region_df = pd.json_normalize(tmp)
    
    # 생성된 모든 컬럼을 .values를 사용하여 할당
    cols_to_assign = ['행정동', '행정동코드', '법정동명', '법정동코드']
    data[cols_to_assign] = region_df[cols_to_assign].values

    print(f"행정동 추출 완료. 최종 행정동 결측 수: {data['행정동'].isna().sum()}/{len(data)}")
    
    return data
```

## 행정동 기반 물재생센터 매핑 함수

```python
import pandas as pd

def map_to_center(df: pd.DataFrame, mapping_path: str, save_path: str = None, df_name: str = "인허가_df"):
    """
    df와 매핑 파일을 이용해 '물재생센터'를 병합하는 함수.
    
    Parameters
    ----------
    df : pd.DataFrame
        원본 데이터프레임 (행정동, 행정동코드 포함 필요).
    mapping_path : str
        매핑 엑셀 파일 경로.
    save_path : str, optional
        결과 저장 경로. None이면 저장하지 않음.
    df_name : str, default="인허가_df"
        결과 파일명 prefix.
    
    Returns
    -------
    merged : pd.DataFrame
        매핑 완료된 데이터프레임
    """
    # 매핑 데이터 불러오기
    mapping_df = pd.read_excel(mapping_path, engine='openpyxl')
    
    # 행정동명 정정
    name_fix = {
        '신설동': '용신동', 
        '용두동': '용신동', 
        '면목제3.8동': '면목3.8동'
    }
    df = df.copy()
    df['행정동'] = df['행정동'].replace(name_fix)
    
    # 이름 기준 병합 시도 함수
    def try_merge_by_name(df_in, map_in):
        if map_in['H_DNG_NM'].duplicated().any():
            return None
        merged_name = df_in.merge(
            map_in[['H_DNG_NM', '처리센터']],
            how='left',
            left_on='행정동',
            right_on='H_DNG_NM',
            validate='m:1'
        ).rename(columns={'처리센터':'물재생센터'}).drop(columns=['H_DNG_NM'])
        return merged_name
    
    # 첫 번째 병합 시도
    merged = try_merge_by_name(df, mapping_df)

    if merged is None:
        # 중복 진단
        g = mapping_df.groupby('H_DNG_NM')['처리센터'].nunique().reset_index(name='센터_고유값_수')
        conflict_keys = g.loc[g['센터_고유값_수'] > 1, 'H_DNG_NM'].tolist()

        if len(conflict_keys) == 0:
            map_dedup_safe = mapping_df.drop_duplicates(subset=['H_DNG_NM', '처리센터'], keep='first')
            if map_dedup_safe['H_DNG_NM'].duplicated().any():
                need_fallback = True
            else:
                need_fallback = False
                merged = try_merge_by_name(df, map_dedup_safe)
        else:
            need_fallback = True

        # 코드 기준 폴백
        if merged is None and need_fallback:
            df['행정동코드'] = pd.to_numeric(df['행정동코드'], errors='coerce').astype('Int64')
            df['행정동코드_trim'] = (df['행정동코드'] // 100).astype('Int64')

            mapping_df['행정동코드'] = pd.to_numeric(mapping_df['행정동코드'], errors='coerce').astype('Int64')

            ys_code = mapping_df.loc[mapping_df['H_DNG_NM'] == '용신동', '행정동코드']
            if ys_code.empty:
                raise ValueError("매핑 파일에서 '용신동'의 행정동코드를 찾지 못했습니다.")
            ys_code = ys_code.iloc[0]

            df.loc[df['행정동'].eq('용신동'), '행정동코드_trim'] = ys_code

            merged = df.merge(
                mapping_df[['행정동코드', '처리센터']].drop_duplicates(subset=['행정동코드']),
                how='left',
                left_on='행정동코드_trim',
                right_on='행정동코드'
            ).rename(columns={'처리센터':'물재생센터'})
    
    # 결과 요약 출력
    print(f"[요약] 전체 행: {len(merged):,d}")
    print(f"[요약] 매핑 성공: {(~merged['물재생센터'].isna()).sum():,d}")
    print(f"[요약] 매핑 실패: {merged['물재생센터'].isna().sum():,d}")
    
    # 저장 옵션 처리
    if save_path is not None:
        full_path = f"{save_path}/{df_name}_물재생센터매핑_결과.csv"
        merged.to_csv(full_path, index=False, encoding='utf-8-sig')
        print(f"[저장 완료] {full_path}")
    
    return merged
```

## 일자별, 물재생센터별 단순 운영숫자 카운트 함수


```python
### 단순하게 물재생센터별로 카운트만 할거라면,,
# %% [markdown]
# # 일자×물재생센터×업종별 일별 카운트 생성 함수 + 파일명에 데이터 태그 포함 (Py 3.8+ 호환)
# - 저장 시 파일명에 인허가 데이터 이름(예: "목욕장업", "수영장업")을 포함하기 위해 `data_tag` 파라미터 추가
# - 예: data_tag="목욕장업" → 인허가_목욕장업_일자별_센터×업종_시설수_20170101_20250630.csv

# %%
import re
import pandas as pd
from itertools import product
from typing import Optional, List  # 3.8+ 호환 타입

def _sanitize_tag(s: str) -> str:
    """
    파일명 안전 문자열로 변환:
    - 양쪽 공백 제거
    - 공백은 '_'로
    - 슬래시 등 위험 문자는 '-'로 치환
    """
    s = s.strip()
    s = s.replace(" ", "_")
    # 파일명에 흔히 문제되는 문자들 제거/치환
    s = re.sub(r"[\\/:*?\"<>|]+", "-", s)
    return s

def build_daily_center_industry_counts(
    df: pd.DataFrame,
    out_path: Optional[str] = None,
    start: str = "2017-01-01",
    end: str = "2025-06-30",
    center_col: str = "물재생센터",
    industry_col: str = "개방서비스명",
    close_col: str = "폐업일자",
    centers_override: Optional[List[str]] = None,     # 예: ["서남","탄천","중랑","난지"]
    industries_override: Optional[List[str]] = None,  # 특정 업종 집합 강제 사용 시 지정
    data_tag: Optional[str] = None,                   # ✅ 파일명에 넣을 인허가 데이터 이름(예: "목욕장업")
    date_tag_fmt: str = "%Y%m%d",                     # 파일명 날짜 포맷
) -> pd.DataFrame:
    """
    지정한 기간 동안 매일 `물재생센터 × 업종`별 운영 시설 수(카운트)를 계산해 반환하고,
    필요 시 CSV로 저장합니다. 저장 파일명에 data_tag(인허가 데이터 이름)를 포함할 수 있습니다.

    Returns
    -------
    pd.DataFrame  # 컬럼: ['date', center_col, '업종', '시설수_운영']
    """
    # -------------------------
    # 1) 기간/그리드 기본 준비
    # -------------------------
    START = pd.Timestamp(start)
    END = pd.Timestamp(end)
    date_index = pd.date_range(START, END, freq="D")

    # -------------------------
    # 2) 컬럼 검사 & 날짜 변환
    # -------------------------
    required = {center_col, industry_col, close_col}
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"입력 데이터에 필요한 컬럼 누락: {missing}")

    df = df.copy()
    df[close_col] = pd.to_datetime(df[close_col], errors="coerce")

    # -----------------------------------------
    # 3) 운영기간 계산 (개업일 정보 없다고 가정)
    # -----------------------------------------
    df["영업시작일"] = START
    df["영업종료일_포함"] = (df[close_col] - pd.Timedelta(days=1)).fillna(END)
    df["영업종료일_포함"] = df["영업종료일_포함"].where(df["영업종료일_포함"] <= END, END)
    df = df[df["영업종료일_포함"] >= df["영업시작일"]].copy()

    # -----------------------------------------
    # 4) 시설 → 일자 확장 (센터/업종 보존)
    # -----------------------------------------
    def make_daily_rows(row: pd.Series) -> pd.DataFrame:
        dts = pd.date_range(max(row["영업시작일"], START),
                            min(row["영업종료일_포함"], END),
                            freq="D")
        if len(dts) == 0:
            return pd.DataFrame(columns=["date", center_col, industry_col, "시설_운영"])
        return pd.DataFrame({
            "date": dts,
            center_col: row[center_col],
            industry_col: row[industry_col],
            "시설_운영": 1,   # 이후 sum = 해당 날짜/센터/업종의 시설 수
        })

    parts = [make_daily_rows(r) for _, r in df.iterrows()]
    fac_daily = (pd.concat(parts, ignore_index=True)
                 if parts else
                 pd.DataFrame(columns=["date", center_col, industry_col, "시설_운영"]))

    # -----------------------------------------
    # 5) 날짜×센터×업종 집계
    # -----------------------------------------
    if fac_daily.empty:
        grp = pd.DataFrame(columns=["date", center_col, industry_col, "시설수_운영"])
    else:
        grp = (fac_daily
               .groupby(["date", center_col, industry_col], as_index=False)
               .agg(시설수_운영=("시설_운영", "sum")))

    # -----------------------------------------
    # 6) 풀그리드 생성(날짜 × 센터 × 업종) 후 병합
    # -----------------------------------------
    centers = (centers_override
               if centers_override is not None
               else sorted(df[center_col].dropna().unique().tolist()))
    industries = (industries_override
                  if industries_override is not None
                  else sorted(df[industry_col].dropna().unique().tolist()))

    full_grid = pd.DataFrame(
        list(product(date_index, centers, industries)),
        columns=["date", center_col, industry_col]
    )
    out_df = full_grid.merge(grp, on=["date", center_col, industry_col], how="left")
    out_df["시설수_운영"] = out_df["시설수_운영"].fillna(0).astype(int)

    # -----------------------------------------
    # 7) 정렬 및 반환(업종 별칭)
    # -----------------------------------------
    out_df = (out_df
              .sort_values(["date", center_col, industry_col], ignore_index=True)
              .rename(columns={industry_col: "업종"}))

    # -----------------------------------------
    # 8) 저장(선택) — 파일명에 data_tag 포함
    # -----------------------------------------
    if out_path is not None:
        start_tag = START.strftime(date_tag_fmt)
        end_tag = END.strftime(date_tag_fmt)
        tag_part = f"{_sanitize_tag(data_tag)}_" if data_tag else ""
        fname = f"인허가_{tag_part}_일자별_물재생센터_운영수_{start_tag}_{end_tag}.csv"
        save_full = f"{out_path.rstrip('/')}/{fname}"
        out_df.to_csv(save_full, index=False, encoding="utf-8-sig")
        print(f"[저장 완료] {save_full}")

    # 요약
    print(
        f"[요약] 행수={len(out_df):,d}, 날짜={out_df['date'].nunique():,d}, "
        f"센터={out_df[center_col].nunique():,d}, 업종={out_df['업종'].nunique():,d}"
    )
    return out_df

# %% [markdown]
# ## 사용 예시
# - 데이터셋이 '목욕장업'이면 data_tag="목욕장업" 같이 넘겨 파일명을 구분합니다.
# - 센터명을 4개로 고정하고 싶으면 centers_override에 표준 라벨을 맞춰 주세요.

# %%
# result = build_daily_center_industry_counts(
#     df=merged_pool,                               # 이미 메모리에 있는 DF 사용
#     out_path="../data/processed",
#     start="2017-01-01",
#     end="2025-06-30",
#     center_col="물재생센터",
#     industry_col="개방서비스명",
#     close_col="폐업일자",
#     centers_override=["서남", "탄천", "중랑", "난지"],
#     industries_override=None,
#     data_tag="목욕장업"                           # ✅ 파일명에 포함될 태그
# )
# result.head()
```
