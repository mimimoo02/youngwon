```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/processed/서울_AWS_통합_일별_20170101_20250630_습도완전보완_컬럼삭제.csv', encoding='utf-8-sig')
df
```

```python
df.info()
```

```python
### 불쾌지수 계산 함수 ###

def calc_discomfort_index(T, H):
    """
    T: 기온(℃)
    H: 상대습도(%)
    return: 불쾌지수
    """
    return 0.81 * T + 0.01 * H * (0.99 * T - 14.3) + 46.3

# 예시: 기온 30도, 습도 70%
di = calc_discomfort_index(30, 70)
print(f"불쾌지수: {di:.2f}")
```

```python
df['불쾌지수'] = calc_discomfort_index(df['일_평균기온(°C)'], df['평균습도(%)'])
df
```

```python
### 불쾌지수등급 함수 ###

def discomfort_level(di):
    if di < 68:
        return '쾌적'
    elif di < 75:
        return '약간 불쾌'
    elif di < 80:
        return '불쾌'
    elif di < 84:
        return '매우 불쾌'
    else:
        return '극심한 불쾌'

df['불쾌지수등급'] = df['불쾌지수'].apply(discomfort_level)
df[['일_평균기온(°C)', '평균습도(%)', '불쾌지수', '불쾌지수등급']]
df
```
