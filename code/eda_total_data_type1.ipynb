### 서남  EDA ###
- 하수처리량 컬럼 중 합계만 사용(1처리장, 2처리장, 시설현대화 컬럼 제거) : 합계에 모든 값의 합이 있는거니까
- 연도, 월, 일별 비교를 위해 컬럼에 추가
- 불쾌지수등급별 하수처리량 비교, 불쾌지수 높은 날이랑 강수량이랑도 관계 있는지 파악
- 계절별, 연도별, 월별, 일별, 요일별 하수처리량 비교

```python
# 불필요한 처리장 컬럼 제거
center_s = center_s.drop(columns=['1처리장', '2처리장', '시설현대화'])
```

```python
### 날짜 파생 컬럼 추가 - 연도, 월, 일, 계절, 주말
center_s['연도'] = center_s['날짜'].dt.year
center_s['월'] = center_s['날짜'].dt.month
center_s['일'] = center_s['날짜'].dt.day

# 요일 숫자화 (월:0 ~ 일:6), 주말 여부
center_s['요일(숫자)'] = center_s['날짜'].dt.weekday
center_s['주말'] = center_s['요일(숫자)'].isin([5, 6]).astype(int)

# 계절 파생
def get_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    else:
        return '겨울'

center_s['계절'] = center_s['월'].apply(get_season)
```

```python
center_s.describe()
```

```python
### 불쾌지수 등급별 하수 처리량 비교
# 불쾌지수 등급별 boxplot
sns.boxplot(x='불쾌지수등급', y='합계', data=center_s)
plt.title('불쾌지수 등급별 하수 처리량 분포')
plt.show()
```

```python
### 불쾌지수가 높은 날 vs 강수량 관계
# 불쾌지수 75 이상인 날만 필터링
hot = center_s[center_s['불쾌지수(DI)'] >= 75]

# 산점도
sns.scatterplot(x='불쾌지수(DI)', y='일_일강수량(mm)', data=hot)
plt.title('불쾌지수가 높은 날의 강수량 분포')
plt.show()
```

```python
### 하수처리량 비교(계절, 연도, 월, 일, 요일)
# 계절
sns.boxplot(x='계절', y='합계', data=center_s)
plt.title('계절별 하수 처리량')
plt.show()

# 연도
sns.lineplot(x='연도', y='합계', data=center_s.groupby('연도')['합계'].mean().reset_index())
plt.title('연도별 평균 하수 처리량')
plt.show()

# 월
sns.boxplot(x='월', y='합계', data=center_s)
plt.title('월별 하수 처리량')
plt.show()

# 일
sns.boxplot(x='일', y='합계', data=center_s)
plt.title('일자별 하수 처리량')
plt.show()

# 요일
order = ['월요일','화요일','수요일','목요일','금요일','토요일','일요일']
sns.boxplot(x='요일', y='합계', data=center_s, order=order)
plt.title('요일별 하수 처리량')
plt.show()
```

```python
### 요일별 생활인구 vs 하수처리량
# 요일별 평균 생활인구
sns.barplot(x='요일', y='생활인구', data=center_s, order=order, estimator=np.mean)
plt.title('요일별 평균 생활인구')
plt.show()

# 요일별 평균 하수량
sns.barplot(x='요일', y='합계', data=center_s, order=order, estimator=np.mean)
plt.title('요일별 평균 하수 처리량')
plt.show()
```

```python
### 연도별 계절별 하수 처리량

# 계절-연도별 평균 집계
season_year_avg = center_s.groupby(['연도', '계절'])['합계'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.pointplot(data=season_year_avg, x='연도', y='합계', hue='계절', markers='o', linestyles='-')
plt.title('연도별 계절별 평균 하수 처리량')
plt.ylabel('평균 하수 처리량')
plt.xlabel('연도')
plt.legend(title='계절')
plt.tight_layout()
plt.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```

```python
### 계절별 강수량 비교 - 박스플롯 : 강수량 분포 확인
plt.figure(figsize=(8, 6))
sns.boxplot(data=center_s, x='계절', y='일_일강수량(mm)', palette='pastel')

plt.title('계절별 일일 강수량 분포')
plt.xlabel('계절')
plt.ylabel('일일 강수량 (mm)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
### 계절별 평균 강수량 – 막대 그래프 (Barplot)
plt.figure(figsize=(8, 6))
sns.barplot(data=center_s, x='계절', y='일_일강수량(mm)', estimator='mean', ci='sd', palette='Set2')

plt.title('계절별 평균 일일 강수량')
plt.xlabel('계절')
plt.ylabel('평균 일일 강수량 (mm)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
season_rain = center_s.groupby('계절')['일_일강수량(mm)'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=season_rain, x='계절', y='일_일강수량(mm)', palette='Set3')

plt.title('계절별 총 강수량')
plt.xlabel('계절')
plt.ylabel('총 강수량 (mm)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
